```python
import pandas as pd
import os
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
factor_names = [f'factor_iter{i}.csv' for i in range(5)]
xseq_names = [f'xseq_iter{i}.csv' for i in range(5)]
yseq_name = 'yseq.csv'
```

```python
for f in factor_names:
    df = pd.read_csv(f, index_col="Epoch")
    fig, ax = plt.subplots(1,3, figsize=(16, 4))
    df.plot(y=['Factorization Loss F', 'Factorization Loss X'], ax=ax[0])
    df.plot(y='Temporal Loss X', ax=ax[1])
    df.plot(y='Validation Loss (Global)', ax=ax[2])
```

```python
for f in xseq_names:
    df = pd.read_csv(f, index_col="Epoch")
    df.plot(title=f)
```
